# matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# specifically for jupyter notebook to show graph is notebook replace of plt.show()

```python
%matplotlib inline     
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y,'g') # r is default use for color
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')
plt.show()
```

#  creating multiple plots on same canvas

```python
plt.subplot(1,2,1)   #(1,2,1) first row two columns 1st element
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(2,2,1)   #(1,2,1) first row two columns 1st element
plt.plot(x,y,'r^-')
plt.subplot(2,2,4)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(2,2,1)   #(1,2,1) first row two columns 1st element
plt.plot(x,y,'r^-')
plt.subplot(2,2,2)
plt.plot(y,x,'g*-')
plt.subplot(2,2,3)
plt.plot(y,x,'b*-')
plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
fig=plt.figure()
# main axes
axes1=fig.add_axes([0.1,0.1,0.8,0.8])  #left,ottom,width,height
axes2=fig.add_axes([0.2,0.5,0.4,0.3])   #insiide axis

#larger fig
axes1.plot(x,y,'g--')
axes1.set_xlabel("X-main Fig")
axes1.set_ylabel("Y-main fig")
axes1.set_title("Main Figure")

#small fig
axes2.plot(x,y,'r*-')
axes2.set_xlabel("X-main Fig")
axes2.set_ylabel("Y-main fig")
axes2.set_title("Main Figure")
```

```python
fig=plt.figure()
# main axes
axes1=fig.add_axes([0.9,0.9,0.8,0.8])  #left,bottom,width,height
axes2=fig.add_axes([0.1,0.7,0.4,0.7])   #insiide axis

#larger fig
axes1.plot(x,y,'g--')
axes1.set_xlabel("X-main Fig")
axes1.set_ylabel("Y-main fig")
axes1.set_title("Main Figure")

#small fig
axes2.plot(x,y,'r--')
axes2.set_xlabel("X-main Fig")
axes2.set_ylabel("Y-main fig")
axes2.set_title("Main Figure")
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])   #scale of add_axes is form 0 to 1
ax.plot(x,y,label='x**2',color='g')
ax.plot(x,x**3,label='x**3')
ax.legend(loc=2)  #loc=1,2,3,4
```

# histogram

```python
from random import sample
data=sample(range(1,1000),100)
plt.hist(data,histtype='bar')
# data
```

```python
z=[50,60,70,80,90]
label=['rent','shopping','food','grocerry','personal']
Explode=[0.1,0,0,0,0]
plt.pie(z,labels=label,explode=Explode)
plt.title('Expenses')
plt.legend(title='Expenses')
```

```python
z=[50,60,70,80,90]
label=['rent','shopping','food','grocerry','personal']
Explode=[0.1,0,0,0,0]
plt.bar(label,z)
plt.title('Expenses')
plt.legend(title='Expenses')
```
